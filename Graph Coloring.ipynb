## Теория

Раскраска графа - это функция $f: X \rightarrow \mathbb{N} $, сопоставляющая некоторому множеству элементов $X$ графа, например, множеству вершин или ребер, натуральное число, именуемое цветом. Бывают, соответственно, вершинные, реберные и другие раскраски графа. Когда не упомянают, о какой раскраске идет речь, обычно имеют ввиду вершинную.

Правильной называют такую раскраску, в которой:
1. две смежные вершины имеют разные цвета;
2. число цветов минимально

Число цветов в правильной раскраске называют хроматическим числом графа.

Задача раскраски графа состоит в поиске правильной раскраски графа. Такая задача является NP-полной.

Иногда бывает достаточно узнать лишь хроматическое число графа. Для этого достаточно научиться отвечать на вопрос "Можно ли раскрасить граф в $q$ цветов?". Тогда, начиная с $q=1$, будем увеличивать $q$, пока ответ на этот вопрос не станет положительным. Поскольку $q \leq |V|$, поиск хроматического числа полиномиально сводятся к задаче проверки на раскрашиваемость.

Задачу о раскраске тоже можно свести к задаче проверки на раскрашиваемость за полиномиальное время:

<b>Теорема 1.</b>

Пусть $isColoring(G, q)$ - алгоритм, отвечающий на вопрос, можно ли раскрасить $G$ в $q$ цветов. Пусть $\varphi(n)$ - время его работы.<br>
Тогда алгоритм $getColoring(G)$, выдающий раскраску графа, можно воспроизвести за время $\varphi(n) poly(n)$, где $poly(n)$ - многочлен константной степени, где $n = |V(G)|$

<b>Доказательство:</b>

Приведем алгоритм, сводящий $getColoring(G)$ к $isColoring(G, q)$.

Пусть $q$ - хроматическое число графа $G$. Выше описан способ его получения за $\varphi(n) poly(n)$

Если $G$ - полный граф, красим каждую вершину в свой цвет. Иначе есть хотя бы $2$ несмежные вершины $v$ и $w$. Дополним граф $G$ ребром $(v, w)$. С помощью алгоритма $isColoring(G + (v, w), q)$ проверим, можно ли раскрасить полученный граф в $q$ цветов.

Если этого сделать нельзя, значит для любой раскраски $f$ графа $G$ в $q$ цветов, $f(v) = f(w)$. Рассмотрим сужение $G|_{(v, w)}$ - граф полученный из $G$ "стягиванием" ребра $(v, w)$ в вершину $v$ следующим образом: $\forall w' \in V(G) \backslash \{v\}, (w', w) \in E(G): (w', w) \rightarrow (w', v)$, то есть соседи $w$ становятся соседями $v$. При этом ребро $(v, w)$ и вершина $w$ исчезают. Любая раскраска $f$ сужения $G|_{(v, w)}$ дополненная правилом $f(w) = f(v)$ является раскраской графа $G$. Таким образом мы рекурсивно свели поиск раскраски $G$ к поиску раскраски $G|_{(v, w)}$.

Если $G + (v, w)$ можно раскрасить в $q$ цветов, тогда существует раскраска $G$ в $q$ цветов, в которой $f(v) \not= f(w)$. В этом случае мы рекурсивно свели поиск раскраски $G$ к поиску раскраски $G + (v, w)$

Заметим, что операцивания стягивания, как и добавления ребра не удаляет уже существующих ребер. Поскольку на каждом шаге для пары вершин либо одна из вершин удалялась, либо дополнялась ребром, дважды одна и та же пара быть выбрана не может. Значит глубина рекурсии не превышает число пар $n(n-1)$. На каждом шаге рекурсии задействовался алгоритм $isColoring$. Следовательно время работы полученного алгоритма $\varphi(n) poly(n)$.

Теорема доказана.

Таким образом задача проверки на раскрашиваемость, также является NP-полной.

Реализации таких задач, ввиду класса NP, которому они принадлежит, требуют больших временных ресурсов, что не всегда отвечает требованиям. Тогда можно пытаться извлечь пользу из упрощенных модификаций, например, найти неточное значение хроматического числа или проверить только на раскрашиваемость в 2 цвета. Здесь же подробно рассмотрен алгоритм проверки на раскрашиваемость,  работающий за экспоненциальное время, а также решение задачи о раскраске на его основе.

Напомним, что независимое множество - это такое подмножество вершин графа, между вершинами которых в графе нет ребер. Обозначим $G|_{W}$ - ограничение графа $G$ на множество вершин $W$, т.е. подграф, полученный из $G$ выкидыванием вершин не входящих в $W$ и ребер, из них выходящих.

<b>Теорема 2.</b>

Пусть дан $G(V, E)$.<br>
Определим для $W \subseteq V$, $g(W)$ - число непустых независимых множеств в $G|_{W}$.<br>

Тогда $G$ может быть покрашен в $q$ цветов тогда и только тогда, когда

$$\sum_{W \subseteq V}{(-1)^{|V \backslash W|} (g(W))^q} > 0$$

<b>Доказательство:</b>

$(g(W))^q$ - число способов выбрать $q$ непустых независимых множеств из $W$.
Пусть $U \subseteq V$. Положим $h(U)$ - число способов выбрать $q$ непустых независимых множеств, объединение которых дает $U$.
Следовательно, $(g(W))^q = \sum_{U \subseteq W}{h(U)}$. Тогда

$$\sum_{W \subseteq V}{(-1)^{|V \backslash W|} (g(W))^q} = \sum_{W \subseteq V}{(-1)^{|V \backslash W|} \sum_{U \subseteq W}{h(U)}} = \sum_{U \subseteq V}{h(U)} \sum_{W \supseteq U}{(-1)^{|V \backslash W|}} = \sum_{U \subseteq V}{h(U)} \sum_{W' \subseteq V \backslash U}{(-1)^{|W'|}}$$

Покажем, что для любого непустого множества $X$ число подмножеств нечетного размера $|ODD(X)|$  совпадает с числом подмножеств четного размера $|EVEN(X)|$: зафиксируем произвольный $x \in X$; существует биекция $\alpha(Y): ODD(X) \leftrightarrow EVEN(X), \alpha(Y) = Y \triangle \{x\}$.

Значит, за исключением случая $W' = \emptyset \iff U = W = V$, число положительных слагаемых в сумме $\sum_{W' \subseteq V \backslash U}{(-1)^{|W'|}}$ равно числу нечетных, т.е. сумма равна $0$. Следовательно,

$$\sum_{W \subseteq V}{(-1)^{|V \backslash W|} (g(W))^q} = h(V)$$

Таким образом $\sum_{W \subseteq V}{(-1)^{|V \backslash W|} (g(W))^q}$ - число способов выбрать $q$ непустых независимых множеств, объединение которых дает $V$. Если это число положительное, $G$ может быть покрашен в $q$ цветов по принципу "каждое независимое множество в свой цвет; если вершина лежит в нескольких множествах, красим в любой". Иначе покрасить $G$ в $q$ цветов нельзя.

Теорема доказана.

Для каждого $W \subseteq V$ мы можем посчитать g(W) за время $O(2^{|W|} |E|)$ посредством проверки каждого подмножества $W$ на независимость. Если же нам доступна экспоненциальная от числа вершин память, можно поступить намного лучше.

Обозначим $N(v)$ - множество соседей $v$, т.е. смежных $v$ вершин; $N[v] := N(v) \cup \{v\}$

<b>Теорема 3.</b>

Пусть дан $G(V, E)$, $W \subseteq V$.<br>
Определим $g(\emptyset) = 0$, тогда $\forall v \in W$

$$g(W) = g(W \backslash \{v\}) + g(W \backslash N[v]) + 1$$

где $N[v]$ - множество соседей $v$.

<b>Доказательство:</b>

Зафиксируем произвольную вершину $v \in W$. Разделим все непустые независимые множества в $W$ на те, которые содержат $v$ и не содержат $v$. Число последних равно $g(W \backslash \{v\})$.

Рассмотрим независимые множества содержащие $v$. Так как множества независимые, они не содержат $N(v)$. Покажем, что их число совпадает с числом независимых (не обязательно непустых) множеств в $W \backslash N[v]$: существует биекция, сопоставляющая любому независимому множеству $S$, $\{v\} \subseteq S \subseteq W$, независимое множество $S' \subseteq W \backslash N[v]$ по правилу $S \rightarrow S \backslash \{v\} = S'$.

Число непустых независимых множеств в $W \backslash N[v]$ равно g(W \backslash N[v]). Независимому множеству $\{v\}$ сопоставлено $\emptyset$. Таким образом, число независимых множеств, содержащих $v$ равно $g(W \backslash N[v]) + 1$.

Теорема доказана.

Теперь, мы можем быстро вычислять очередное $g(W)$ на основе уже посчитанных $g(W')$, где $W' \subset W$.

Итак, алгоритм включений-исключений состоит из 2х шагов:

1. Положим $g(\emptyset) = 0$. Для каждого подмножества $W$ множества вершин $V$ вычислим $g(W)$ по формуле $g(W) = g(W \backslash \{v\}) + g(W \backslash N[v]) + 1$ и запомним его;
2. Посчитаем $h = \sum_{W \subseteq V}{(-1)^{|V \backslash W|} (g(W))^q}$, результатом будет $(h>0)$.

Так как подмножеств всего $2^{|V|}$, оба шага алгоритма требуют по $2^{|V|} poly(|V|)$ времени и в совокупности столько же памяти. Следовательно, алгоритм включений-исключений имеет время работы $2^{n} poly(n)$ и требует $2^{n} poly(n)$ места, что можно записать с использованием асимптотических обозначений в виде $O(2^n)$ времени и $O(2^n)$ памяти, где $n$ - число вершин данного графа.

## Реализация

Для работы с графами, воспользуемся библиотекой <a href="http://networkx.github.io">networkx</a>. Воспользуемся функцией генерации случайного графа, чтобы получить тестовый пример. Количество вершин и вероятность возниконовения ребра можно регулировать.

```python
import networkx as nx

verticies_num = 4
probability = 0.6

G=nx.fast_gnp_random_graph(verticies_num, probability)

print(G.nodes())
print(G.edges())
```

Для визуализации графов networkx интегрирована с библиотекой <a href="http://matplotlib.org/">matplotlib</a>.

```python
%matplotlib inline
# чтобы изображения выводились прямо в ipython notebook, а не в отдельном окне

import matplotlib.pyplot as plt

nx.draw_circular(G, node_size=700)
# построили граф, расположив вершины по кругу

nx.draw_networkx_labels(G,pos=nx.circular_layout(G), font_size=20)
# наложили на них нумерацию

plt.show()
# вывели на экран
```

Для перебора всех подмножеств мы будем использовать двоичные маски. К примеру маска $10100$ задает подмножество, состоящее из $1$ и $3$ элемента. Если записать маски в десятичном виде, то последовательность $0, 1, ... , 2^{|V|}$ задает все подмножества $V$. Научимся переводить числа в двоичный вид:

```python
i = 23
print bin(i), "двоичный вид"
print bin(i)[2:], "двоичный вид без лишних символов"
print bin(i)[2:][::-1], "реверс предыдущего"
```

Для удобства напишем функцию, возвращающую множество вершин по маске:

```python
def get_tuple(mask):
    i = 0
    t = ()
    for bit in bin(mask)[2:][::-1]:
        if bit == '1':
            t += (G.nodes()[i], )
        i += 1
    return t

i = 11
print bin(i)[2:][::-1]
print get_tuple(i)
```

Теперь мы готовы написать рекурсивную функцию $g(W)$, вычисляя ее по формуле $g(W) = g(W \backslash \{v\}) + g(W \backslash N[v]) + 1$.

```python
g_cache = dict({(): 0})

def g(subset):
    if subset in g_cache:
        return g_cache[subset]
    else:
        v = subset[0]
        v_neighbors = list(nx.all_neighbors(G, v))
        subset_without_v = subset[1:]
        subset_without_v_and_neighbors = tuple(x for x in subset_without_v if x not in v_neighbors)
        g_cache[subset] = g(subset_without_v) + g(subset_without_v_and_neighbors) + 1
        return g_cache[subset]
```

Эта функция кэширует полученные значения в контейнере $g\_cache$, где ключом является подмножество $W$, а значением - $g(W)$. Чтобы завершить шаг 1 алгоритма включений-исключений, запустим $g(subset)$ на всех подмножествах $V$, чтобы точно заполнить весь $g\_cache$.

```python
for i in range(1, 2 ** G.number_of_nodes()):
    g(get_tuple(i))


# выведем отсортированный g_cache, если он не займет много места на экране
if len(g_cache) <= 16:
    print 'g(W) W'
    for i in sorted(g_cache.items(), key=lambda t: len(t[0])):
        print i[1], ' ', i[0]
```

Остался второй шаг - подсчитать сумму $h = \sum_{W \subseteq V}{(-1)^{|V \backslash W|} (g(W))^q}$ для некоторого $q$:

```python
q = 3

h = 0
for i in range(1, 2 ** G.number_of_nodes()):
    W = get_tuple(i)
    h += ((-1) ** (G.number_of_nodes() - len(W))) * (g_cache[W] ** q)

print (h > 0)
```

Выведенное значение означает, можно ли раскрасить граф в $q$ цветов.

Соберем все в единую функцию:

### Алгоритм включений-исключений

```python
def inclusion_exclusion_algorithm(G, q):
    def get_tuple(mask):
        i = 0
        t = ()
        for bit in bin(mask)[2:][::-1]:
            if bit == '1':
                t += (G.nodes()[i], )
            i += 1
        return t

    g_cache = dict({(): 0})
    
    def g(subset):
        if subset in g_cache:
            return g_cache[subset]
        else:
            v = subset[0]
            v_neighbors = list(nx.all_neighbors(G, v))####
            subset_without_v = subset[1:]
            subset_without_v_and_neighbors = tuple(x for x in subset_without_v if x not in v_neighbors)
            g_cache[subset] = g(subset_without_v) + g(subset_without_v_and_neighbors) + 1
            return g_cache[subset]
    
    def get_g():
        for i in range(1, 2 ** G.number_of_nodes()):
            g(get_tuple(i))

    def get_h():
        h = 0
        for i in range(1, 2 ** G.number_of_nodes()):
            W = get_tuple(i)
            h += ((-1) ** (G.number_of_nodes() - len(W))) * (g_cache[W] ** q)
        return h

    get_g()
    return get_h() > 0
```

С его помощью можно посчитать хроматическое число:

### Подсчет хроматического числа

```python
def chromatic_number(G):
    q = 1
    while not inclusion_exclusion_algorithm(G, q):
        q += 1
    return q
```

```python
print chromatic_number(G)
```

Теперь реализуем алгоритм, описанный в Теореме 1, чтобы получить раскраску:

### Поиск раскраски графа

```python
def graph_coloring(G):
    q = chromatic_number(G)
    
    def reqursive_coloring(G):
        if G.number_of_nodes() * (G.number_of_nodes() - 1) == 2 * G.number_of_edges():
        # если граф полный
            coloring = dict()
            color = 0
            for v in G:
                coloring[v] = color
                color += 1
            return coloring
        else:
            for v1 in G.nodes():
                for v2 in G.nodes():
                    if v1 != v2 and (v1, v2) not in G.edges() and (v2, v1) not in G.edges():
                        H = G.copy()
                        H.add_edge(v1, v2)
                        if inclusion_exclusion_algorithm(H, q):
                            return graph_coloring(H)
                        else:
                            for neighbor in nx.all_neighbors(G, v2):
                                if (v1, neighbor) not in G.edges() and (neighbor, v1) not in G.edges():
                                    H.add_edge(v1, neighbor)
                            H.remove_node(v2)
                            coloring = graph_coloring(H)
                            coloring[v2] = coloring[v1]
                            return coloring
    
    return reqursive_coloring(G)
```

```python
f = graph_coloring(G)
print f
```

Таким образом мы получили множество пар, где ключ - номер вершины, значение - ее цвет. Упорядочим его и сформируем массив цветов, где на $i$-й позиции стоит цвет $i$-й вершины:

```python
f = sorted(f.items())
print f
colors = [color for key, color in f]
print colors
```

Повторим рисование графа, только теперь зададим цвета:

```python
nx.draw_circular(G, node_size=700, node_color=colors)
nx.draw_networkx_labels(G,pos=nx.circular_layout(G), font_size=20)
plt.show()
```
